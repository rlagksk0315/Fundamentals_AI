```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(iris)

X = iris.data[:, :]
y = iris.target
y_name = iris.target_names
```

```python
X_train = []
y_train = []
X_test = []
y_test = []

for i in range(0, len(X)):
    if (i % 15 == 14):
        X_test.append(X[i])
        y_test.append(y[i])
    else:
        X_train.append(X[i])
        y_train.append(y[i])
```

```python
import numpy as np

class KNN_Classifier: # K-Nearest Neighbors Classifier
    def __init__(self, K):
        self.K = K # number of neighbors
        
    def fit(self, x_train, t_train): # training set 학습시키기
        self.x_train = x_train
        self.t_train = t_train
    
    def cal_distance(self, point, data): # 두 data 간의 Euclidean distance 구하기
        distance = np.sqrt(np.sum(np.square(point - data)))
        return distance
        
    def majority_vote(self, distances): # data가 속할 class 정할 때 필요한 majority vote
        distances = np.array(distances)
        distances_sorted = distances.argsort()
        distances[distances_sorted][::-1]
                
        votes = {}
        
        for idx in distances_sorted[:self.K]:
            if self.t_train[idx] in votes:
                votes[self.t_train[idx]] += 1
            else:
                votes[self.t_train[idx]] = 1

        maj_vote = max(votes, key = votes.get) # 투표값이 제일 큰 key 값 불러오기
        
        return int(maj_vote) # 투표 결과의 target 값 반환
    
    def predict(self, x_test): # 예측값 구하기 --> 무슨 클래스(label)에 속해있는지
        t_predict = []

        for x in x_test:
            distance_result = []
            for i in range(len(self.x_train)):
                train_distance = self.cal_distance(self.x_train[i], x) # training data - testing data 거리 구하기
                distance_result.append(train_distance) # 구한 eucledian distance를 리스트에 저장
            
            maj_vote = self.majority_vote(distance_result)
            t_predict.append(maj_vote)
        
        return t_predict
```

```python
KNN3 = KNN_Classifier(K = 3)
KNN3.fit(X_train, y_train)

y_pred_3 = KNN3.predict(X_test)
        
for i in range(len(X_test)):
    print('Test Data Index: {0} Computed Class: {1} True Class: {2}'.format(i, y_name[y_pred_3[i]], y_name[y_test[i]]))
```

```python
KNN5 = KNN_Classifier(K = 5)
KNN5.fit(X_train, y_train)

y_pred_5 = KNN5.predict(X_test)

for i in range(len(X_test)):
    print('Test Data Index: {0} Computed Class: {1} True Class: {2}'.format(i, y_name[y_pred_5[i]], y_name[y_test[i]]))
```
